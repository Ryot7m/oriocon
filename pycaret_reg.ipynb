```python
"""
 ロジスティック回帰を使用した検索キーワードのに対する精度評価
 """
import numpy as np
import pandas as pd
import pycaret.classification as pc
import pycaret
print(pycaret.__version__)
```

```python
df_house = pd.read_csv("7-4-18_data.tsv",sep = "\t",header = 0)
two_q = df_house.iloc[:, 208:228] #検索キーワード("Q25"全て)
site_q = df_house.iloc[:, 239:241] #使用サイトと実際の行動("Q28","Q29",Q30)
q_sq = df_house["SQ7"]
q = df_house["Q32"] #最も使用した情報サイト
log_q = pd.concat([two_q, site_q, q_sq], axis = 1)
log_q = log_q.dropna()
log_q = log_q.astype("int")
log_qst = log_q.replace({"Q29":[3,4,5,6,7,10,13], "SQ7": [3,4,5,6,7,10,13]}, 28)
log_word = log_qst.iloc[:, :20]
log_q
```

```python
un_set = pc.setup(log_word, target = log_q["Q29"])
```

```python
best = un_set.create_model("lightgbm")
```

```python
uni_param = un_set.tune_model(best, optimize = "Accuracy") #ハイパーパラメータの最適化を行う
uni_param
```

```python
un_set.plot_model(uni_param, plot = "parameter") #パラメーターの確認
```

```python
un_set.plot_model(uni_param, plot = "auc") #ROC曲線の結果
```

```python
un_set.plot_model(uni_param, plot = "feature")
```
